```python
import pandas as pd
import nltk
```

```python
data = pd.read_csv('texts.csv').set_index('site')
```

```python
data['freqDist'] = data.text.apply(lambda text: nltk.FreqDist([t for t in text.split()]))
```

```python
for site in data.index:
    print(site)
    data.loc[site].freqDist.plot(5, cumulative=False)
```

```python
from nltk.corpus import stopwords
```

```python
sw = stopwords.words('german') + stopwords.words('english')
```

```python
def clean(text):
    # remove non-alphanumeric characters (e.g. "(", ":", ...)) except spaces and punctuation
    text = ''.join([t for t in text if t in [" ",".",",",";"] or t.isalpha()])
    
    # "tokenize" text: "This is a sentence, ok" -> ["This", "is", "a", "sentence", "ok"]
    text = nltk.word_tokenize(text, "german")
    
    # remove stopwords (and, or, ...), convert to lowercase, exclude really short words
    text = [t.lower() for t in text if t not in sw and len(t)>2]
    
    return text
```

```python
data['cleanText'] = data.text.apply(clean)
```

```python
data['cleanFreqDist'] = data.cleanText.apply(lambda text: nltk.FreqDist(text))
```

```python
for site in data.index:
    print(site)
    data.loc[site].cleanFreqDist.plot(5, cumulative=False)
```
